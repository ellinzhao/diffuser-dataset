```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt
from scipy import ndimage
import myfunctions as funcs
```

```python
# ---------------------- #
#  Function definitions  #
# ---------------------- #


class TwoDImg:
    
    def __init__(self, n, r, k):
        self.n = n
        self.r = r
        self.k = k
        self.arr, self.centers = self.gen_2d_points(n, r, k)

    def gen_2d_points(self, n, r, k):
        """
        n = dim of the image.
        r = max radius of point.
        k = num of random points.
        all params >= 1.
        """
        arr = np.ones((n, n))
        centers = []
        while k > 0:
            a = np.random.randint(0, n)
            b = np.random.randint(0, n)
            centers += [(a, b)]
            r1 = np.random.randint(r/2, r)
            # (a, b) is center of point, r1 is radius for point. 

            y, x = np.ogrid[-a:n-a, -b:n-b]
            mask = x*x + y*y <= r1*r1
            arr[mask] = 255
            k -= 1
        return arr, centers
    
    def get_img(self):
        # not filtered array
        return self.arr
    
    def show_img(self):
        img_filtered = ndimage.gaussian_filter(self.arr, sigma=7, order=0)
        plt.imshow(img_filtered, cmap='gray', interpolation='nearest');
        
    def get_centers(self):
        return self.centers
    
    def plt_x_section(self):
        img_filtered = ndimage.gaussian_filter(self.arr, sigma=7, order=0)
        x, y = self.centers[0]
        start = -1
        end = -1
        prev = -1
        prev = 100
        for i in range(0, 500):
            curr = img_filtered[x][i]
            if start == -1 and prev == 1.0 and curr != 1.0:
                start = i
            elif start != -1 and end == -1 and prev != 1.0 and curr == 1.0:
                end = i
                break
            prev = curr
        plt.plot(img_filtered[x][start:end])
        plt.show()
```

```python
img = TwoDImg(500, 15, 20)
img.show_img()
```

```python
img.plt_x_section()
```

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt
from scipy import ndimage


def gen_3d_pt(n, r):
    """
    n = dim of the image.
    r = max radius of point.
    k = num of random points.
    all params >= 1.
    """
    a = np.random.randint(0, n)
    b = np.random.randint(0, n)
    c = np.random.randint(0, n)
    r1 = np.random.randint(r/2, r) 
    print(a, b, c, r1)
    arr = np.ones((n, n, n))
        
        # finding the center
    x, y, z = np.ogrid[-a:n-a, -b:n-b, -c:n-c]
        
        # creating mask with radius r1-i
    mask = x*x + y*y + z*z <= (r1)*(r1)
        
    arr[mask] = 255
    return arr


def gen_3d_points(n, r, k):
    """
    n = dim of the image.
    r = max radius of point.
    k = num of random points.
    all params >= 1.
    """
    arr = np.zeros((n, n, n))
    print("done initializing")
    for _ in range(k):
        a = np.random.randint(n/4, 3*n/4)
        b = np.random.randint(n/4, 3*n/4)
        c = np.random.randint(n/4, 3*n/4)
        r1 = np.random.randint(r/2, r) 
        print(a, b, c, r1)
        for i in range(0, r1 + 1):
            x, y, z = np.ogrid[-a:n-a, -b:n-b, -c:n-c]
            # creating mask with radius r1-i
            mask = x*x + y*y + z*z <= (r1-i)*(r1-i)
            arr[mask] = 1
        print("done making 1 point")
    return arr
```

```python
# generate single 3d point
vol = gen_3d_pt(50, 7)
vol_filtered = ndimage.gaussian_filter(vol, sigma=5, order=0)
```
